# Testing

```python
#Here we are importing sklearn

from sklearn.model_selection import train_test_split
X = data.drop(0,axis=1)
y=data[0]
```

```python
#Here we are considering two parameters X and y and 
#traing and testing the data

X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.30)
```

# Decision Tree

The algorithm we use here is Decision Tree :

```python
#Importing DecisionTreeClassifier from sklearn

from sklearn.tree import DecisionTreeClassifier
```

```python
#Creating an Instance

dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

# Prediction

```python
#Here we are predicitong the data

predicitons = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

# Result

```python
print(classification_report(y_test,predicitons))
```

From the result we can say that male and female has same ratio of the disease criteria but one more thing we have to show is the 
age, at which age they were affected more and under which category

```python
print(confusion_matrix(y_test,predicitons))
```

```python
#Here we are using boxplot
sns.boxplot(x=0,y=8,data=data)
```
